```python
a = input()
brackets = []
bracket_1 = a.find('(')
while bracket_1 != -1:
  bracket_2 = a.find(')', bracket_1 + 1)
  if bracket_2 != -1:
    brackets.append(a[bracket_1 + 1: bracket_2])
  bracket_1 = a.find('(', bracket_1 + 1)
print(brackets)
```
